```python
#Dimension Reduction
gender = pd.get_dummies(auto['Gender'], drop_first = True)
```

```python
auto = pd.concat([auto, gender], axis = 1)
auto
```

```python
region_typ = pd.get_dummies(auto['Education Level'], drop_first = True)
region_typ
```

```python
auto = pd.concat([auto, region_typ], axis = 1)
auto
```

```python
# smoker = pd.get_dummies(auto['Years of Experience'], drop_first = True)
# smoker
```

```python
auto.drop(['Gender','Education Level'], axis = 1, inplace = True)
auto
```

```python
#Starting Machine Learning 
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
df_train, df_test = train_test_split(auto, train_size = 0.7, test_size = 0.3, random_state = 1)
```

```python
df_train
```

```python
df_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler
```

```python
df_train.head()
```

```python
# Apply scaler() to all the columns except the 'dummy' variables
num_vars = ['Age', 'Years of Experience', 'Salary']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
# Let's check the correlation coefficients to see which variables are highly correlated

plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True,)
plt.show()
```

```python
y_train = df_train.pop('Salary')
X_train = df_train
```

```python
y_train
```

```python
X_train
```

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
# lm.fit(X_train, y_train)

#rfe = RFE(lm, 10)
rfe = RFE(lm, n_features_to_select=10)
#running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
X_train.columns
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
X_train_rfe
```

```python
# Adding a constant variable
import statsmodels.api as sm
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
X_train_rfe
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()  # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

```python
df_test
```

```python
num_vars = ['Age', 'Years of Experience', 'Salary']

df_test[num_vars] = scaler.transform(df_test[num_vars])
```

```python
df_test
```

```python
y_test = df_test.pop('Salary')
X_test = df_test
```

```python
y_test
```

```python
# Now let's use our model to make predictions.

# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[['Age','Years of Experience','Male']]

X_test_new
```

```python
# Adding a constant variable
X_test_new = sm.add_constant(X_test_new)
X_test_new
```

```python
# Making predictions
y_pred = lm.predict(X_test_new)
y_pred
```

```python
y_test
```

```python
#test aucc
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)              # Plot heading
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16)                          # Y-label
```
